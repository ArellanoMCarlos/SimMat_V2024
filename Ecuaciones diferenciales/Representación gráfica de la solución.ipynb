## Solución en el tiempo
El modelo del sistema `masa-resorte` está descrito por la siguiente **ecuación diferencial**:

\begin{equation}
\frac{d^2x}{dt^2} + \frac{k}{m}x = 0,
\end{equation}

cuya solución se escribe como
\begin{equation}
x(t) = A \cos(\omega_{o} t) + B \sin(\omega_{o} t)
\end{equation}
Y su primera derivada (velocidad) sería
\begin{equation}
v(t) = \frac{dx(t)}{dt} = \omega_{0}[- A \sin(\omega_{0} t) + B\cos(\omega_{0}t)]
\end{equation}

**¿Cómo se ven las gráficas de $x$ vs $t$ y $\frac{dx}{dt}$ vs $t$?**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A=1
B=1
def x(t,w0):
  return A*np.cos(w0*t)+B*np.sin(w0*t)
def v(t,w0):
  return w0*(-A*np.sin(w0*t)+B*np.cos(w0*t))
```

```python
# Forma 1 de definir subgráficas
fig,ax=plt.subplots(2,1)#,sharex=True)
W0=np.linspace(0.001,2,10)
t=np.linspace(0,10,500)
for w0 in W0:
  # Posiciones
  ax[0].plot(t,x(t,w0),label='$\omega_0={}$'.format(round(w0,3)))
  # Velocidades
  ax[1].plot(t,v(t,w0),label='$\omega_0={}$'.format(round(w0,3)))

ax[0].grid()
ax[1].grid()

plt.legend(loc='center left',bbox_to_anchor=(1,1.1)) # loc=location
```

```python
# Forma 2
fig=plt.figure()
fig.suptitle('Sistema masa-resorte')

# Posiciones
ax=fig.add_subplot(2,1,1)
for w0 in W0:
  ax.plot(t,x(t,w0),label='$\omega_0={}$'.format(round(w0,3)))
ax.grid()
ax.set(ylabel='(m)',title='Posición')
ax.label_outer()

# Velocidades
ax=fig.add_subplot(2,1,2)
for w0 in W0:
  ax.plot(t,v(t,w0),label='$\omega_0={}$'.format(round(w0,3)))
ax.grid()
ax.set(xlabel='Tiempo (s)',ylabel='(m/s)',title='Velocidad')

plt.legend(loc='center left',bbox_to_anchor=(1,1.1))
```

```python
# Forma 3
fig=plt.figure()
gs=fig.add_gridspec(2,1,hspace=0.5) # Especificamos la posición de los subplots
ax=gs.subplots()
fig.suptitle('Sistema masa-resorte')

for w0 in W0:
  # Posiciones
  ax[0].plot(t,x(t,w0),label='$\omega_0={}$'.format(round(w0,3)))
  # Velocidades
  ax[1].plot(t,v(t,w0),label='$\omega_0={}$'.format(round(w0,3)))

ax[0].grid()
ax[0].set(ylabel='(m)',title='Posición')
ax[1].grid()
ax[1].set(xlabel='Tiempo (s)',ylabel='(m/s)',title='Velocidad')

plt.legend(loc='center left',bbox_to_anchor=(1,1.1)) # loc=location
```

## Gráfico interactivo
La posición y velocidad para el sistema `masa-resorte` se escriben como:
\begin{align}
x(t) &= x(0) \cos(\omega_{0} t) + \frac{\dot{x}(0)}{\omega_{0}} \sin(\omega_{0} t)\\
\dot{x}(t) &= -\omega_{0}x(0) \sin(\omega_{0} t) + \dot{x}(0)\cos(\omega_{0}t)
\end{align}

**Animación del péndulo.**

La solución para el sistema reducido del `péndulo` es

\begin{align}
\theta(t) &= \theta(0) \cos(\omega_{0} t) + \frac{\dot{\theta}(0)}{\omega_{0}} \sin(\omega_{0} t)\\
\dot{\theta}(t) &= -\omega_{0}\theta(0) \sin(\omega_{0} t) + \dot{\theta}(0)\cos(\omega_{0}t).
\end{align}

Las posiciones $x$ y $y$ se obtienen de la descomposición entre la longitud y el ángulo:

\begin{align}
x &= l \sin(\theta), \quad
y = -l \cos(\theta) .
\end{align}
