```python
# Librer√≠as
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Inizializando la figura
fig,ax = plt.subplots(subplot_kw={'projection':'3d'})

# Creando los datos x,y,z
z=np.linspace(-5,5,1000)
r=z**2+1
th=np.linspace(-8*np.pi,8*np.pi,1000)
x=r*np.cos(th)
y=r*np.sin(th)

# La gr√°fica
ax.plot(x,y,z)
```
