```python
sym.var('x',real=True)
def A(x):
  return x*(6-2*x)
dA=A(x).diff(x)
c=sym.solve(dA)
c
```

```python
def S(x):
  y=36/x**2
  return 4*x**2+6*x*y
dS=S(x).diff(x)
c=sym.solve(dS)
c
```

```python
def L(x):
  h1=(5**2+x**2)**(1/2)
  h2=(4**2+(2.2-x)**2)**(1/2)
  return h1+h2
dL=L(x).diff(x)
c=sym.solve(dL)
c
```
