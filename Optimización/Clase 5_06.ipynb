```python
import numpy as np
import scipy.optimize as opt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
A=np.random.rand(17)
B=np.random.rand(17)
C=np.random.rand(17)
datos=pd.DataFrame()
datos['A']=A
datos['B']=B
datos['C']=C
datos['Yield']=np.random.rand(17)
```

```python
def f(x,A,B,C):
  return x[0]+x[1]*A+x[2]*B+x[3]*C+x[4]*A*C+x[5]*C**2
def fun(x,A,B,C,Yield):
  return np.mean((np.sqrt(Yield)-f(x,A,B,C))**2)
```

```python
x0=np.random.rand(6)
sol=opt.minimize(fun,x0,args=(datos['A'],datos['B'],datos['C'],datos['Yield']))
```

```python
for v in ['A','B','C']:
  plt.figure()
  plt.plot(datos[v],np.sqrt(datos['Yield']),'o')
  plt.plot(datos[v],f(sol.x,datos['A'],datos['B'],datos['C']),'o')
  plt.grid()
```
